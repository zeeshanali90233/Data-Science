## Numpy vs Pandas

NumPy is best for numerical calculations, while Pandas is ideal for data cleaning, manipulation, and tabular data handling.  
Both work together, but Pandas adds structure and convenience for real-world datasets.

## What is Pandas?

Pandas is a Python library for data manipulation and analysis.

### DataFrame
like Excel tables (rows & columns)

```python
import pandas as pd
```

```python
# Series: 1D data
data = [10, 20, 30, 40]
s = pd.Series(data)
print(s)
```

```python
# DataFrame: 2D table
data = {
    'Name': ['Ali', 'Sara', 'Zeeshan'],
    'Age': [25, 22, 28],
    'City': ['Lahore', 'Karachi', 'Islamabad']
}

df = pd.DataFrame(data)
print(df)
```

```python
# Access a column
print(df['Name'])

```

```python
# Access a column
print(df['City'])

```

```python

# Access multiple columns
print(df[['Name', 'City']])
```

```python

# Access rows by index
print(df.loc[0])    # first row
```

```python
print(df.iloc[1])   # second row
```

```python
# Statistics
print(df['Age'].mean())   # average age
print(df['Age'].sum())    # sum of ages
```

```python
print(df[df['Age'] > 23])   # rows where Age > 23
```

```python
# Add a new column
df['Country'] = 'Pakistan'
print(df)
```

```python

# Write DataFrame to CSV
df.to_csv('output.csv', index=False)
```

```python
# Read CSV file
df = pd.read_csv('output.csv')
```
