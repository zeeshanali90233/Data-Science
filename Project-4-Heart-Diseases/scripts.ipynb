### We feel there is issue with ploting count plot because the more number of records will incluence one side

```python
# We will Go for Percentages

df_group = df.groupby(['Heart_Disease', 'Gender']).size()
print(df_group)
```

```python
# Making it more readable dataframe 

df_group = df_group.reset_index(name='Count')
print(df_group)
```

```python
# Heart Disease by Age Range
df['Age_Group'] = pd.cut(df['Age'], bins=[0,30,40,50,60,70,80,100])
df.groupby(['Age_Group', 'Heart_Disease']).size().reset_index(name='Count')
```

```python
# Heart Disease by BMI Category

df['BMI_Group'] = pd.cut(df['BMI'], bins=[0,18.5,25,30,35,60],
                         labels=['Underweight','Normal','Overweight','Obese','Extremely Obese'])
df.groupby(['BMI_Group','Heart_Disease']).size().reset_index(name='Count')
```
