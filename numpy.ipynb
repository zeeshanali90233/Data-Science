```python
heights=np.array([1,2,3,4])
print("Datatype of Array: ", heights.dtype)
```

```python
floatArray=heights.astype(float)
print("Datatype of Array: ", floatArray.dtype)
```

#math operations

```python
heights=np.array([1,2,3,4])

print(np.pow(heights,2))
print(np.multiply(heights,3))
print(np.mod(heights,3))
```

```python
print("Min: ",np.min(heights))
print("Max: ",np.max(heights))
print("SQRT: ",np.sqrt(heights))
print("sin: ",np.sin(heights))
print("cos: ",np.cos(heights))
```

#Reshaping

```python
heights=np.array([1,2,3,4,5,6])

reshaped=np.reshape(heights,(3,2))
print(reshaped)
```

# Indexing Slicing

```python
heights=np.array([1,2,3,4])

print(heights[2])
print(heights[1:4])
print(heights[::-1])
print(heights[:4])
print(heights[2:])
```

#Iteration

```python
for i in heights:
  print(i)
```

#Copy

```python
arr1=np.array([1,2,3,4])
arr2=arr1

arr2[2]=34

# This also changed arr1
print(arr1)
print(arr2)
```

```python
arr1=np.array([1,2,3,4])
arr2=arr1.copy()

arr2[2]=34

# This also changed arr1
print(arr1)
print(arr2)
```

#Concatenate

```python
arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])

arr3=np.concatenate((arr1,arr2))

print(arr3)
```

#Delete

```python
# Delete by index

arr1=np.array([1,2,3,4])
arr=np.delete(arr1,2)
print(arr)
```

#Multi Dimensional Arrays in Numpy


```python
oneDim=np.array([1,2,3,4,5])
twoDim=np.array([[1,2,3],[2,3,4]])
threeDim=np.array([[[1,2],[2,3]],[[2,3],[3,3]]])

print(oneDim)
print("Dimension: ",oneDim.ndim)
print(twoDim)
print("Dimension: ",twoDim.ndim)
print(threeDim)
print("Dimension: ",threeDim.ndim)
```
