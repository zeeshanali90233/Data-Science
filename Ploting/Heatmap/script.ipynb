### Heatmap

üîç They show patterns easily ‚Äî help visualize relationships or correlations between variables at a glance.

üé® They highlight intensity or frequency ‚Äî colors make it easy to spot high and low values quickly.

üìà They make large data readable ‚Äî useful for comparing many variables or categories in a compact form.

```python
import pandas as pd 

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Housing.csv')
```

```python
df.head()
```

```python
# 2. Select only 'price' and 'area' columns
df_small = df[['price', 'area']]

# 3. Calculate the correlation between them
corr = df_small.corr()

# 4. Draw the heatmap
plt.figure(figsize=(4, 3))
sns.heatmap(corr, annot=True, cmap='YlGnBu')
plt.title("Heatmap: Price vs Area")
plt.show()
```

```python
# 2. Select only 'price' and 'bedrooms' columns
df_small = df[['price', 'bedrooms']]

# 3. Calculate the correlation between them
corr = df_small.corr()

# 4. Draw the heatmap
plt.figure(figsize=(4, 3))
sns.heatmap(corr, annot=True, cmap='YlGnBu')
plt.title("Heatmap: Price vs Bedrooms")
plt.show()
```
