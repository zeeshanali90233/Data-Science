### What is a Box Plot?
A Box Plot (also called a Box-and-Whisker Plot) shows how data is spread and helps identify outliers.

It displays:

 1. Median (middle value)

 2. Quartiles (Q1, Q3)

 3. Minimum & Maximum values (within range)

 4. Outliers (points beyond whiskers)

We have to use it for 

1. See data spread (how spread out values are)
2. Compare distributions between groups & Detect Outliers

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simple numeric data
data = np.array([12, 15, 14, 10, 18, 22, 30, 25, 14, 16, 12, 11])

# Create Box Plot
plt.boxplot(data)

plt.title("Simple Box Plot Example")
plt.ylabel("Values")
plt.show()
```

### Removing Outliers 

For this the method is used first find interquartile range

IQR= Q3-Q1

Min= Q1-1.5*IQR

Max- Q3+1.5*IQR

Remove all which does not come in between this 


```python
# Calculate IQR
Q1 = prices.quantile(0.25)
Q3 = prices.quantile(0.75)
IQR = Q3 - Q1
print("Q1:", Q1)
print("Q3:", Q3)
```

```python
# Define lower & upper bounds
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Lower Bound:", lower)
print("Upper Bound:", upper)
```

```python
# Filter data
filtered_prices = prices[(prices >= lower) & (prices <= upper)]

plt.boxplot(filtered_prices)
plt.title("House Price Distribution (Without Outliers)")
plt.ylabel("Price")
plt.show()
```
