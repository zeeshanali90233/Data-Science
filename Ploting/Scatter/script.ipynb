### Scatter Plot 

```python
import pandas as pd
```

```python
df=pd.read_csv('Housing.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

# Scatter Plot of Price vs Area
plt.scatter(df['area'], df['price'] / 1e6, color='blue', alpha=0.5)
plt.title('Scatter Plot of Price vs Area')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (Million $)')
plt.show()
```

```python
peoples=pd.read_csv('People.csv')
peoples.head()
```

```python
# Weight vs Height Scatter Plot
plt.scatter(peoples['Weight(kg)'], peoples['Height(cm)'], color='green', alpha=0.5)

plt.title('Weight vs Height Scatter Plot')
plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.show()
```
