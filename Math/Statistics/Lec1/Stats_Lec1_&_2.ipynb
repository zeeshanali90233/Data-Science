```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data = [1, 2, 2, 3, 4, 7, 9, 10, 10, 10]

# Mean, Median, Mode
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True)[0][0]

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
plt.plot(data)
plt.show()
```

```python
# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)

# Add vertical lines for Mean, Median, Mode
plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f"Mean = {mean:.2f}")
plt.axvline(median, color='green', linestyle='--', linewidth=2, label=f"Median = {median:.2f}")
plt.axvline(mode, color='orange', linestyle='--', linewidth=2, label=f"Mode = {mode:.2f}")

# Labels and title
plt.title("Histogram with Mean, Median, and Mode")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
np.random.seed(42)

# Right-skewed (long tail on right)
right_skew = np.random.exponential(scale=2, size=1000)

# Left-skewed (long tail on left)
left_skew = -np.random.exponential(scale=2, size=1000) + 10

# Symmetric (normal)
symmetric = np.random.normal(loc=5, scale=1, size=1000)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Right Skewed
axes[0].hist(right_skew, bins=30, color='skyblue', edgecolor='black')
axes[0].set_title("Right Skewed (Mean > Median > Mode)")

# Symmetric
axes[1].hist(symmetric, bins=30, color='lightgreen', edgecolor='black')
axes[1].set_title("Symmetric (Mean = Median = Mode)")

# Left Skewed
axes[2].hist(left_skew, bins=30, color='salmon', edgecolor='black')
axes[2].set_title("Left Skewed (Mean < Median < Mode)")

plt.tight_layout()
plt.show()
```

```python
print("Right Skewed Skewness:", stats.skew(right_skew))
print("Symmetric Skewness:", stats.skew(symmetric))
print("Left Skewed Skewness:", stats.skew(left_skew))
```

```python
dataset=pd.read_csv("Housing.csv")
```

```python
dataset.head()
```

```python
prices=dataset['price']
```

```python
mean = prices.mean()
median = prices.median()
mode = prices.mode()[0]

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
plt.figure(figsize=(10,6))
plt.hist(prices, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# Vertical lines for Mean, Median, Mode
plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f"Mean = {mean:,.0f}")
plt.axvline(median, color='green', linestyle='--', linewidth=2, label=f"Median = {median:,.0f}")
plt.axvline(mode, color='orange', linestyle='--', linewidth=2, label=f"Mode = {mode:,.0f}")

# 5 Labels and title
plt.title("Price Distribution with Mean, Median, and Mode", fontsize=14)
plt.xlabel("Price", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

### Mean > Median > Mode

### Right Skewed Data means max houses prices are from 3 to 4 million

### SD , Percentiles, and Quartiles

```python

# 2 Basic Stats
mean = prices.mean()
median = prices.median()
mode = prices.mode()[0]
std = prices.std()
var = prices.var()
range_ = prices.max() - prices.min()
```

```python
# 3 Percentiles & Quartiles
q1 = prices.quantile(0.25)
q2 = prices.quantile(0.50)  # median
q3 = prices.quantile(0.75)
iqr = q3 - q1

percentiles = np.percentile(prices, [10, 25, 50, 75, 90])
```

```python
# 4 Skewness & Kurtosis
skew = stats.skew(prices)
kurtosis = stats.kurtosis(prices)

# Print all results
print(" DESCRIPTIVE STATISTICS")
print(f"Count: {len(prices)}")
print(f"Mean: {mean:,.2f}")
print(f"Median: {median:,.2f}")
print(f"Mode: {mode:,.2f}")
print(f"Standard Deviation: {std:,.2f}")
print(f"Variance: {var:,.2f}")
print(f"Range: {range_:,.2f}")
print(f"Q1 (25th): {q1:,.2f}")
print(f"Q2 (50th): {q2:,.2f}")
print(f"Q3 (75th): {q3:,.2f}")
print(f"IQR: {iqr:,.2f}")
print(f"Skewness: {skew:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
print("\nPercentiles (10, 25, 50, 75, 90):", percentiles)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 4))
sns.boxplot(x=prices, color='skyblue', width=0.4)

plt.title(" Box Plot  Price Distribution", fontsize=14, weight='bold')
plt.xlabel("Prices (PKR)", fontsize=12)
plt.grid(True, axis='x', linestyle='--', alpha=0.6)
plt.show()
```

| Measure           | Meaning                     | Interpretation                                                                               |
| ----------------- | --------------------------- | -------------------------------------------------------------------------------------------- |
| **Range**         | Max  Min                   | Tells the total price span between cheapest and most expensive items.                        |
| **Variance / SD** | Average deviation from mean | High SD = prices vary widely; low SD = prices are more consistent.                           |
| **IQR (Q3Q1)**   | Middle 50% spread           | If large, data has big variation in mid-range prices; small IQR = prices are tightly packed. |


| Quartile      | Percentile                    | Meaning                  |
| ------------- | ----------------------------- | ------------------------ |
| **Q1 (25th)** | 25% below this value          | Lower-priced range       |
| **Q2 (50th)** | 50% below this value (Median) | Midpoint of prices       |
| **Q3 (75th)** | 75% below this value          | Upper-middle price range |


| Measure                         | Interpretation                               |
| ------------------------------- | -------------------------------------------- |
| **Skewness > 0**  Right-Skewed | Long tail on right  few expensive prices.   |
| **Skewness < 0**  Left-Skewed  | Long tail on left  few very low prices.     |
| **Skewness  0**  Symmetric    | Prices evenly distributed.                   |
| **Kurtosis > 0**  Leptokurtic  | Sharp peak and heavy tails  many outliers.  |
| **Kurtosis < 0**  Platykurtic  | Flatter distribution  fewer extreme values. |
